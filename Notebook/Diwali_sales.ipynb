# Exploratory Data Analysis (EDA)
Trends, Relations, Distributions haru find out garne. Basically data kasto xa, values haru kasta xan, Columns haru kasta xan vanne kura analaysis garne. Aauta EDA, data cleaning vandha aagadi nai garxam, data cleaning garepaxi aarko EDA garxam. First EDA bata hamile data ko size, null value xa xaina, duplicate xa xaina etc vanne barema thaha pauxam. Ani tespaxi data cleaning garxam

## What is done in EDA
Core things done in EDA after cleaning
1. Understand distributions
You study how each variable behaves.
    Are values skewed or symmetric?
    Are most customers spending low or high?
    Are there extreme values?
(Example: histogram of Amount)

2. Detect outliers (real ones, not errors)
Now outliers represent real behavior, not bad data.
    Big spenders
    Rare but important events
(Boxplots are common here)

3. Study relationships between variables
You ask:
    Does age affect spending?
    Do certain categories generate more revenue?
    Is there correlation between variables?
(Scatter plots, correlations, groupby)

4. Segment the data
You break data into meaningful groups:
    Gender-wise spending
    Age group analysis
    Region / category / product segments
This is where business insights start forming.

5. Validate assumptions
You check if what you expect matches reality.
    Are festivals increasing sales?
    Are repeat customers spending more?

6. Prepare for modeling or reporting
Final EDA ensures:
    Features are usable
    Patterns are consistent
    Results are explainable

```python
df.groupby(["Gender"])["Amount"].sum()
# as_index=True by default. Gropuby turns the grouping column into the index. So in this case gender column becomes index, data looks likes series
```

```python
# Groups the data by Gender, calculates the total Amount spent by each gender, and then orders the genders from highest to lowest total spending.
gender_spend = df.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
# Gender stays as a normal column. Output is clean Dataframe
```

```python
ax = sns.barplot(x="Gender", y="Amount", data=gender_spend) # hamile data = gender_spend garna paryo kina vane, sns.barplot() le group tw garxa tara sum chahi gardaina. Vannu ko meaning gender lai male rw female ma group garyo
# tara amount lai group anusar add tw garna paryo, ho yo chahi gardaina. Yesle baru mean nikalxa according to group.
for bars in ax.containers: 
    ax.bar_label(bars) 
```

From above graph we can conclude that female spends more on diwali slaes as compared to male

```python
ax = sns.countplot(x="Age Group", data = df, hue = "Gender")
# Age group lai ajhai Gender ko basis ma divide garxa
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
age_spend = df.groupby(["Age Group", "Gender"], as_index=False)["Amount"].sum().sort_values(by = "Amount", ascending = False)
age_spend
```

```python
ax = sns.barplot(x="Age Group", y = "Amount", hue="Gender", data=age_spend, palette="inferno")
for bars in ax.containers:
    ax.bar_label(bars, fmt='%.0f')
# This code creates a bar plot showing the total Amount spent by each Age Group, 
# split into Male and Female bars
```

From above graph we can conclude that female from age group 26-35 spends most of the amounts in shopping.

```python
Order_numbers_by_state = df.groupby(["State"], as_index=False)["Orders"].count().sort_values(by = "Orders", ascending=False)

sns.set(rc={"figure.figsize":(15, 15)}) # Height(y) = 20, width(x) = 15

ax = sns.barplot(x="State", y = "Orders", data = Order_numbers_by_state, palette="viridis", hue="State") # remove palette and insert  color="red", hue="State" because to use palette we have to set hue parameter, and hue parameter yadi kunai xaina vane hue = x 

for bars in ax.containers:
    ax.bar_label(bars, fmt="%.0f")

plt.xticks(rotation=90)
ax.set_title("Number of Orders by State", fontsize=16)
ax.set_xlabel("State")
ax.set_ylabel("Number of Orders")
ax.grid(True)
plt.show()
```

People from Uttar Pradesh orders the most in diwali

```python
amount_by_state = df.groupby(["State"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)

sns.set(rc={"figure.figsize":(15,15)})
ax = sns.barplot(x="State", y="Amount",hue="State", palette="viridis", data=amount_by_state)


for bars in ax.containers:
    ax.bar_label(bars, fmt="%.0f")
    
plt.xticks(rotation = 90)
ax.set_title("Amount spend by people on each state")
ax.set_ylabel("Amount")
ax.set_xlabel("States")
ax.grid(True)
plt.show()
```

From the figure, we can conclude that people from Uttar Pradesh spend the highest total amount during Diwali. Although Kerala ranks 8th in terms of the number of orders placed, it ranks 10th in total amount spent. This indicates that customers from Kerala place a relatively higher number of orders, but the average value per order is lower compared to other states.

# Conclusion

The analysis shows that **female customers**, especially those aged **26–35**, are the most active and highest spenders during Diwali. **Uttar Pradesh** leads in both the number of orders and total spending, with **female customers from the state** contributing the most. Although **Kerala** places a relatively high number of orders, its total spending is lower, indicating smaller average order values. **Married customers**, particularly married females, and **IT professionals** spend more compared to other groups. While **Clothing and Apparel** receives the highest number of orders, **Food products generate the highest total revenue**, showing higher spending per order.
